```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df['year_month'] = df['date'].str[:7]
df['usage'] = df['pickup'] + df['dropoff']
df = df[['station', 'year_month', 'usage']]
```

```python
df = df.groupby(['year_month', 'station']).sum().reset_index()
```

```python
df
```

```python
location = pd.read_csv('location.csv')
location[['station', 'latitude', 'longitude']]
# location = location.drop('Unnamed: 0')
```

```python
df = pd.merge(df, location,  how='left', left_on=['station'], right_on = ['station'])
```

```python
df
```

```python
df.to_csv('heat.csv', index=False)
```
